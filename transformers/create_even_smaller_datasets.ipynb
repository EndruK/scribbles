```python
import os
import pickle as pkl
import random
from tqdm import tqdm
import multiprocessing
import math
```

```python
file_list_dump = "file_list.pkl"
with open(file_list_dump, "rb") as f:
    file_list = pkl.load(f)
len(file_list)
```

```python
samples_per_file = int(1e4)
```

```python
random.shuffle(file_list)
```

```python
output_folder = "/mnt/media/Corpora/extern/Github_Java_Corpus/transformers/split_corpus"
```

```python
processes = 7
complete_length = len(file_list)
print("amount of files in corpus = %d " % complete_length)
loops = int(complete_length / samples_per_file)
print("amount of loops = %d" % loops)
print("num of processes = %d" % processes)

loops_per_process = loops / processes
print("loops per process = %f" % loops_per_process)

over_sampled = round(math.remainder(loops_per_process, int(loops_per_process)) * processes)

distribution = [math.floor(loops_per_process)] * processes
for i in range(0, over_sampled):
    distribution[i] += 1

indices = []
cnt = 0
for i, e in enumerate(distribution):
    prev_sum = sum(distribution[:i])
    offset = prev_sum * samples_per_file
    process_numbers = []
    for j in range(e):
        start = offset + j * samples_per_file
        process_numbers.append([cnt, start, start + samples_per_file])
        cnt += 1
    indices.append(process_numbers)
for i in indices:
    for j in i:
        print(j)
    print()
```

```python
def process_func(complete_list, index_array, process_id):
    print("start process %d" % process_id)
    for file_index, start, end in index_array:
        file_name = "%03d.java_github_10k.raw" % file_index
        fqn = os.path.join(output_folder, file_name)
        output_file = open(fqn, "a")
        error = 0
        failed_files = []
        for i in range(start, end):
            try:
                with open(complete_list[i], "r") as f:
                    text = f.read()
                output_file.write(text)
                output_file.write("\n<|endoftext|>\n")
            except Exception as e:
                error += 1
                failed_files.append([complete_list[i], type(e)])
        output_file.close()
        
    print("finished process %d" % process_id)
```

```python
def lol_process():
    for i in range(0, int(1e15)):
        i * i ** i ** i
```

```python
process_list = []
pool = multiprocessing.Pool(processes=processes)
for process_id in range(processes):
    process = pool.apply_async(lol_process, args=())
    process_list.append(process)
```

```python
process_list = []
pool = multiprocessing.Pool(processes=processes)
for process_id in range(processes):
    process = pool.apply_async(process_func, args=(file_list, indices[process_id], process_id))
    process_list.append(process)
```
