```python
import os
import pickle as pkl
import random
import math
from tqdm import tqdm
```

```python
path_to_corpus = "/mnt/media/Corpora/extern/Github_Java_Corpus/java_projects"
path_to_result = "/mnt/media/Corpora/extern/Github_Java_Corpus/transformers/smaller_unified_corpus"
```

```python
paths_file = "file_list.pkl"
with open(paths_file, "rb") as f:
    file_list = pkl.load(f)
```

```python
len(file_list)
### sample 1million files
sample_files = random.sample(file_list, int(5e5))
```

```python
split = [.8, .12, .08]
random.shuffle(sample_files)
train_len = math.floor(len(sample_files)*split[0])
valid_len = math.floor(len(sample_files)*split[1])
train = sample_files[:train_len]
valid = sample_files[train_len:train_len+valid_len]
test = sample_files[train_len+valid_len:]
```

```python
def export(paths, name):
    errors = 0
    cnt = 0
    export_file = open(os.path.join(path_to_result, name), "a")
    for p in tqdm(paths):
        try:
            with open(p, "r", encoding='utf8') as f:
                text = f.read()
            export_file.write(text + "\n")
            export_file.write("<|endoftext|>\n")
        except Exception as e:
            print("failed to read file %s" % p)
    export_file.close()
    print("failed files: %d" % errors)
    print("successful files: %d" % cnt)
```
